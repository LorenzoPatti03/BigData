1. Splittare training set e test set
2. encoding dei dati categorici --> tutti dati numeri
3. Imputazione ( Pulizia del dataset )
4. Feature Selection

```python
import pandas as pd
from sklearn.model_selection import train_test_split

df = pd.read_csv("./heart_disease_uci.csv",sep=';')
df.drop(columns=['id'],inplace=True)
```

```python


#1.split training set e test set

X = df.drop(columns=['num'])
y = df.num

X_tr, X_te , y_tr, y_te = train_test_split( X,y,test_size=0.2, random_state=49, stratify=df['dataset'] )

```

```python
#encoding

from sklearn.preprocessing import OrdinalEncoder

numeric_features = X_tr.select_dtypes(include=['number']).columns.tolist()
categorical_feature = [ col for col in X_tr.columns if col not in numeric_features ]

encoder = OrdinalEncoder()

encoder.set_output(transform='pandas') # forzo sklearn a restituire un dataframe pandas inceve di una matrice, funziona anche senza

X_tr[categorical_feature] = encoder.fit_transform( X_tr[categorical_feature] )
X_te[categorical_feature] = encoder.transform( X_te[categorical_feature] )


```

```python
#Imputazione

from sklearn.impute import SimpleImputer




#impostiamo i valori nan al valore della media statistica corretta a secondo del tipo di feauture

for layer in X_tr['dataset'].unique():
    
    #keep_empty_features nel caso in cui per quella colonna e per quel dataste non ci siano valori non elimina la feature
    
    imputer = SimpleImputer(strategy='median',keep_empty_features=True)
    
    X_tr.loc[ X_tr['dataset'] == layer , numeric_features ] = imputer.fit_transform(
        X_tr.loc[X_tr['dataset'] == layer, numeric_features]
    )
    
    X_te.loc[ X_te['dataset'] == layer , numeric_features ] = imputer.transform(
        X_te.loc[X_te['dataset'] == layer, numeric_features]
    )
    
    imputer = SimpleImputer(strategy='most_frequent',keep_empty_features=True)
    
    X_tr.loc[ X_tr['dataset'] == layer , categorical_feature ] = imputer.fit_transform(
        X_tr.loc[X_tr['dataset'] == layer, categorical_feature]
    )
    
    X_te.loc[ X_te['dataset'] == layer , categorical_feature ] = imputer.transform(
        X_te.loc[X_te['dataset'] == layer, categorical_feature]
    )
    
    print(X_tr)
    print(X_te)
        
```

```python
# 1. Import corretti
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import pandas as pd


scaler = StandardScaler()
X_tr_scaled = scaler.fit_transform(X_tr) 

pca = PCA(n_components=0.90)
X_pca = pca.fit_transform(X_tr_scaled)

print(f"Feature originali: {X_tr.shape[1]}")
print(f"Nuovi componenti PCA che spiegano il 90% della varianza: {X_pca.shape[1]}")
```
