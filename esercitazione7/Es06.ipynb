Basandovi sugli esempi visti a lezione provate qualche query sulle collections di sample_mflix.

```python
db = client.sample_mflix
col = db.movies

# cerca i film rated R con Meryl Streep oppure Charlize Theron
query = {"$or" : [{"cast":"Meryl Streep"}, {"cast" : "Charlize Theron"}], "rated":"R"}
prop = {"_id":0, "title":1, "year":1, "cast":1, "rated":1}
res = col.find(query, prop)
list(res.limit(3))
```

```python
# cerca i film con Charlize Theron con anno <= 2015
query = {"cast" : "Charlize Theron", "year":{"$lte":2015}}
prop = {"_id":0, "title":1, "year":1, "cast":1, "rated":1}
res = db.movies.find(query, prop)
list(res.limit(3))
```

```python
# cerca i film che non sono di type movie e hanno num_mflix_comments > 0
query = {"type": {"$ne" : "movie"}, "num_mflix_comments": {"$gt" : 0}}
prop = {"_id":0, "title":1, "type":1, "cast":1, "num_mflix_comments":1}
res = db.movies.find(query, prop)
list(res.limit(3))
```

```python
# cerca i film che hanno un imdb rating > 9
query = {"imdb.rating": {"$gt" : 9}}
prop = {"_id":0, "title":1, "type":1, "cast":1, "imdb":1}
res = db.movies.find(query, prop)
list(res.limit(3))
```

# Esercizio 2
Create un dataframe con il "fullplot" di tutti i film per cui è disponibile.

Basandovi su un criterio di similarità fra testi a vostra scelta trovate i 5 film con la trama più simile al film "Casablanca"

```python
fullplot = db.movies.find({"fullplot":{"$exists":"true"}}, {"_id":0, "title":1, "year":1, "runtime":1, "fullplot":1})
df_fullplot = pd.DataFrame(fullplot)


```

```python

import pandas as pd
import Levenshtein



best = pd.DataFrame(columns=["title", "distance"])

casablanca_fullplot = df_fullplot[df_fullplot["title"] == "Casablanca"].iloc[0]["fullplot"]

for row in df_fullplot.itertuples():
    if row.title == "Casablanca":
        continue
    
    distance = Levenshtein.distance(row.fullplot, casablanca_fullplot)
    
    new_row = pd.DataFrame([{"title": row.title, "distance": distance}])
    best = pd.concat([best, new_row], ignore_index=True)

best = best.sort_values(by="distance").reset_index(drop=True)

# Aggiungi similarità normalizzata (0=diverso, 1=identico)
max_distance = best["distance"].max()
best["similarity"] = 1 - (best["distance"] / max_distance)


print(best.head(5))

    
    
    
```

# Esercizio 3
Utilizzando un dataframe basato sulla collection comments determinate i 5 utenti che hanno fatto più commenti.

```python
import pandas as pd

db = client["sample_mflix"]
com = db["comments"]

com_df = pd.DataFrame(com.find())

# conteggio dei commenti per utente (metodo più semplice ed efficiente)
commenti_utente = com_df['name'].value_counts().to_dict()

order = dict(sorted(commenti_utente.items(), key=lambda x: x[1], reverse=True))

# mostra i primi 5 utenti con il relativo numero di commenti
top5 = list(order.items())[:5]
print(top5)
```

1. Trovare tutti gli alloggi di tipo 'House' in United States che possono ospitare 4 persone

```python
query = { "property_type" : "House", "address.country" : "United States", "beds": 4 } 
projection =  { "_id": 0 , "property_type":1, "address.country":1, "beds":1, "name":1}
res = lar.find(query,projection)   
print(list(res))
```

2. Trovare gli alloggi che hanno un numero di review maggiore di 100

```python

exp = { 
    "$expr": { 
        "$gt": [ { "$size": "$reviews" }, 100 ] 
    }
}

proj = { "_id": 0, "name": 1 }

res = lar.find(exp, proj)
print(list(res))
```

3. Trovare gli alloggi che hanno fra 3 e 5 letti

```python
query = {"beds": {"$gte": 3, "$lte": 5}}
proj = { "_id": 0, "name": 1, "beds": 1 }
res = lar.find(query, proj)
print(list(res))
```

4. Trovare gli allogi che hanno fra le amenities "Wifi" e "Microwave"

```python

query = {"amenities": {"$all": ["Wifi", "Microwave"]}}
proj = {"_id": 0, "name": 1, "amenities": 1}
res = lar.find(query, proj)
print(list(res))

```

6. Trovare il numero di alloggi in "United States"

```python
total_us_listings = lar.count_documents({"address.country": "United States"})
print("Numero di alloggi in United States:", total_us_listings)

#mettere match prima per abbattere il data processing
#match_stg = { "$match" : { "address.country": "United States" } }
group_stg = { "$group" : { "_id" : "$address.country", "total_listings": { "$sum": 1 } } }
match_stg = { "$match" : { "_id": "United States" } }

res = lar.aggregate([group_stg, match_stg])
list(res)
```

7. Trovare il numero di alloggi con più di 2 letti per ciascun paese

```python


match = { "$match" : {  "beds": { "$gt": 2 } } }
grp = { "$group" : { "_id": "$address.country" , "total_listings": { "$sum": 1 } } }

res = lar.aggregate([match, grp])
list(res)

```

8. Trovare i 5 alloggi in United States col minor numero di amenities

```python
match = { "$match" : { "address.country": "United States"}}
limit = { "$limit" : 5 }

pipeline = [
    match,
    {"$project": {
        "_id": 0,
        "name": 1,
        "amenities_count": {"$size": "$amenities"}
    }},
    {"$sort": {"amenities_count": 1}},
    limit
]




res = lar.aggregate(pipeline)
list(res)
```
