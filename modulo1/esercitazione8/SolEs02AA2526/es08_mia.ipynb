```python
#!pip install neo4j

from neo4j import GraphDatabase

# Update with your Neo4j credentials and connection URI
uri = "bolt://54.197.28.170"
user = "neo4j"
password = "catches-glues-instructor"

driver = GraphDatabase.driver(uri, auth=(user, password))
```

```python
def run_query(query):
    with driver.session() as session:
        result = session.run(query)
        return [record.data() for record in result]
```

```python
#Movies released after 2000

query = """
    MATCH (m:Movie)
    WHERE m.released > 2000
    RETURN m.name, m.released
"""

run_query(query)
```

```python
#Actors who acted in The Matrix

query = """ 
    MATCH (p:Person)-[:ACTED_IN]->(m: Movie { title: "The Matrix"})
    RETURN p.name, m.title
"""

run_query(query)
```

```python
# People who directed and acted in movies

query = """ 
    MATCH (p:Person)-[:ACTED_IN]->(m:Movie)<-[:DIRECTED]-(p:Person)
    RETURN p
"""

run_query(query)


#soluzione prof: prende i registi che hanno almeno recitato in un film, io prendo un regista che ha registato nello stesso film che ha diretto
query3 = """
MATCH (p:Person)-[:DIRECTED]->(m:Movie)
WHERE (p)-[:ACTED_IN]->()
RETURN p.name AS director_actor, m.title AS movie
"""

run_query(query3)
```

```python
#Actor movie counts

query = """ 
    MATCH (p:Person)-[:ACTED_IN]->(m:Movie)
    RETURN p.name, count(m) AS n_film
"""

run_query(query)
```

```python
#Actor pairs who acted in the same movie

query = """ 
    MATCH (p1:Person)-[:ACTED_IN]->(m:Movie)<-[:ACTED_IN]-(p2:Person)
    WHERE p1 <> p2
    RETURN p1.name, p2.name, m.title
"""

run_query(query)
```
